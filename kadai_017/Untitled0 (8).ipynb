```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です")
    else :
      print(f"{self.name}は大人ではありません")

human1 = Human("山中", 30)
human2 = Human("佐藤", 24)
human3 = Human("伊藤", 10)

human_list = [human1, human2, human3]

for human in human_list:
  human.check_adult()





```
